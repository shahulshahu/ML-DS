# Importing Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore', category=RuntimeWarning)
```

```python
from sklearn.datasets import load_breast_cancer
data_cancer = load_breast_cancer()
data_cancer
```

```python
df_cancer = pd.DataFrame(np.c_[data_cancer['data'], data_cancer['target']], 
                         columns=np.append(data_cancer['feature_names'], ['target']))
df_cancer.head()
```
